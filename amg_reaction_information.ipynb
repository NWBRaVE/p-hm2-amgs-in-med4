```python
import json
import cobra
```

```python
model = cobra.io.read_sbml_model("model_files/ProchlorococcusMED4v1.xml")
amgs = json.loads(open("model_files/amg_reactions.json").read())
manual_relabels = json.loads(open("model_files/subsystem_alias.json").read())
amg_gene_names = json.loads(open("model_files/amg_gene_map.json").read())
amg_rxn_names = {k: f"{v}: {k}" for k, v in amg_gene_names.items()}
```

```python
for amg in sorted(amgs, key=amg_rxn_names.get):
    reaction = model.reactions.get_by_id(amg)
    ss = manual_relabels[reaction.subsystem]
    met_names = {}
    for mets in reaction.metabolites:
        met_names[mets.id] = f"[{mets.name}]"
    rxn_str = str(reaction)
    for m, n in met_names.items():
        rxn_str = rxn_str.replace(m, n)
    if rxn_str.endswith("<=> "):
        rxn_str = rxn_str + "⌀"
    rxn_str = rxn_str.replace("<=>", r"\rightleftharpoons\newline")
    rxn_str = rxn_str.replace("-->", r"\harpoon\newline")
    rxn_str = amg_rxn_names[amg] + " &" + rxn_str.split(":")[1]
    print(f"{rxn_str} & {ss}" + r"\\\hline")
```
